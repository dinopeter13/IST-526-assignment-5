#(Bonus) Visualization combining time-series and text data
In this part, I would like to know number of words in article
when time goes by. First, I compute number of words for each article
and make them as a list, finally, I add them as a new column in
original time series dataframe and made a visualization.

```python
df_time = pd.read_csv('data/news_articles.csv', parse_dates=True, index_col = "Date",encoding='cp1252')
df = pd.read_csv('data/news_articles.csv', encoding='latin-1')

new_list = []
for column in df['Article']:
  word_list = column.split()
  number_of_words = len(word_list)
  new_list.append(number_of_words)

df_time['number_of_words'] = new_list

df_time['Date'] = df_time.index



fig = px.line(
    data_frame = df_time, 
    x='Date', 
    y='number_of_words', 
    title='number of words for in different articles'
)

fig.show()

```
